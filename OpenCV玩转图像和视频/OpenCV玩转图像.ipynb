# Opencv 读取图片

```python
cat_img = cv2.imread("./img/cat.jpg")
```

```python
type(cat_img)
```

```python
cat_img_fixed = cv2.cvtColor(cat_img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(cat_img_fixed)
```

```python
cat_gray = cv2.imread("./img/cat.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
cat_gray.shape
```

# OpenCV 实现对图片的压缩

```python
plt.imshow(cat_img_fixed)
```

```python
cat_img_fixed.shape
```

```python
cat_img_resize = cv2.resize(cat_img_fixed, (200, 300))
```

```python
plt.imshow(cat_img_resize)
```

```python
cat_img_resize.shape
```

# OpenCV 实现图像的翻转

```python
# 0代表垂直翻转 1代表水平翻转 -1代表水平垂直都翻转
cat_img_flip = cv2.flip(cat_img_fixed, -1)
```

```python
plt.imshow(cat_img_flip)
```

# OpenCV实现图像的保存

```python
img_save = cv2.cvtColor(cat_img_flip, cv2.COLOR_RGB2BGR)
```

```python
cv2.imwrite("./img/cat_img_flip.jpg", img_save)
```
